### Load and resample the dataset from half hours into hours:

```python
df = pd.read_csv('nyc_taxi.csv')
df['timestamp'] = pd.to_datetime(df['timestamp'])
df.set_index('timestamp', drop=True, inplace=True)
df = df.resample("H").sum()
# df['Date'] = df.index
print('Size of df:',df.shape[0])
df.head(3)
```

```python
## only 2014:
df = df[df.index <= df.index[int(len(df.index)*0.82)]]
```

```python
## feature to predict:
feat = 'value'
```

### Visualise data split between training and test sets:

```python
split_ = df.index[int(len(df.index)*0.8)]

df_train = df.loc[df.index <= split_].copy()
df_test = df.loc[df.index > split_].copy()

_ = df_test \
    .rename(columns={feat: 'TEST SET'}) \
    .join(df_train.rename(columns={feat: 'TRAINING SET'}), how='outer') \
    .plot(figsize=(20,7), title='')
```

## Create time series features

```python
def create_features(df, label=None):
    """
    Creates time series features from datetime index
    """
    df['date'] = df.index
    df['hour'] = df['date'].dt.hour
    df['dayofweek'] = df['date'].dt.dayofweek
    df['quarter'] = df['date'].dt.quarter
    df['month'] = df['date'].dt.month
    df['year'] = df['date'].dt.year
    df['dayofyear'] = df['date'].dt.dayofyear
    df['dayofmonth'] = df['date'].dt.day
    df['weekofyear'] = df['date'].dt.weekofyear
    df['weekday'] = [i.isoweekday() for i in df.index]
    df['weekend'] = [1 if x <= 5 else 2 for x in df['weekday']]
    
    X = df[['hour','dayofweek','quarter','month','year',
               'dayofyear','dayofmonth','weekofyear','weekday','weekend']]
    if label:
        y = df[label]
        return X, y
    return X
```

```python
X, y = create_features(df, label=feat)

X_train, y_train = create_features(df_train, label=feat)
X_test, y_test = create_features(df_test, label=feat)

X_train.head(3)
```

## Create XGBoost model

```python
# !brew install libomp
import xgboost as xgb
from xgboost import plot_importance, plot_tree

# from sklearn.preprocessing import RobustScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.pipeline import make_pipeline, Pipeline
from sklearn.model_selection import train_test_split, cross_val_score, cross_val_predict, GridSearchCV, RandomizedSearchCV, KFold
```

```python
xgb_model = xgb.XGBRegressor(booster='gbtree')

clf = Pipeline([('xgb', xgb_model)])
param_grid = {
              'xgb__nthread': [12],
              'xgb__max_depth': [6,8],
              'xgb__learning_rate': [0.1], #list(np.arange(0.1, 0.4, 0.1))  0.1 best so far
              'xgb__n_estimators': list(range(50, 250, 100)), 
             }  

model = GridSearchCV(estimator=clf, param_grid=param_grid, verbose=False, n_jobs=-1,iid=True, refit=True, cv=10) #verbose=10 => to see it train

model.fit(X_train, y_train)
```

```python
model.best_score_
```

```python
reg_model = model.best_estimator_
reg_model.fit(X_train,y_train)
```

## Error Metrics on training and testing sets:

```python
print('Training set:  MSE: {:20}, RMSE: {:20}, MAE: {:20}, Corr: {:20}'.format(mean_squared_error(y_true=y_train,y_pred=reg_model.predict(X_train)),
                                        np.sqrt(mean_squared_error(y_true=y_train,y_pred=reg_model.predict(X_train))),
                                        mean_absolute_error(y_true=y_train,y_pred=reg_model.predict(X_train)),
                                        np.corrcoef(reg_model.predict(X_train),y_train.values)[0,1]))

print('Test set:      MSE: {:20}, RMSE: {:20}, MAE: {:20}, Corr: {:20}'.format(mean_squared_error(y_true=y_test,y_pred=reg_model.predict(X_test)),
                                                                       np.sqrt(mean_squared_error(y_true=y_test,y_pred=reg_model.predict(X_test))),
                                                                       mean_absolute_error(y_true=y_test,y_pred=reg_model.predict(X_test)),
                                                                       np.corrcoef(reg_model.predict(X_test),y_test.values)[0,1]))
```

- RMSE is the standard deviation of the residuals (prediction errors). Residuals are a measure of how far from the regression line data points are; RMSE is a measure of how spread out these residuals are. In other words, it tells you how concentrated the data is around the line of best fit.

- The MAE (mean absolute error) is easier to interpret: it's literally the mean of the absolute value of the error.

### Mean absolute percentage error gives an easy to interpret percentage showing how off the predictions are:

```python
def mape(targets, predictions): 
    targets, predictions = np.array(targets), np.array(predictions)
    return str(np.round(np.mean(np.abs((targets - predictions) / targets)) * 100,2))+'%'
```

```python
print('Mean absolute percentage error:  Training: {:7}, Testing: {:7}'.format(mape(df_train[feat],reg_model.predict(X_train)),
                                                                              mape(df_test[feat],reg_model.predict(X_test))))
```

## Forecast on test set

```python
df_test['Prediction'] = reg_model.predict(X_test)
df_all = pd.concat([df_test, df_train], sort=False)
```

```python
# Plot actuals vs forecast
def plot_graph(actuals,pred,start_date,end_date):
    
    fig, ax = plt.subplots(figsize=(20, 7))
    
    _ = df_all[[actuals,pred]].plot(ax=ax,style=['-','-'])
    
    ax.set_xbound(lower=start_date, upper=end_date)
    plot = plt.suptitle('{} - {}: Actuals vs Forecast'.format(start_date,end_date))
    
    return plot;
```
